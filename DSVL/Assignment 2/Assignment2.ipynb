```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('titanic.csv')
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
#dropping unneccessary columns
data.drop(columns = ['Name', 'Ticket', 'Cabin', 'Parch'], axis = 1, inplace = True)
data.head()
```

```python
data.isnull().sum()
```

```python
#handling missing values
age_med = data['Age'].median()
age_med
```

```python
data['Age'].fillna(age_med, inplace = True)
data.isnull().sum()
```

```python
#dropping the rows with null values, for rows whose values cannot be replaced
data.dropna(inplace = True)
data.isnull().sum()
```
