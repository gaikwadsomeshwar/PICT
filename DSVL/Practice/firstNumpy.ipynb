```
# importing the numpy library
import numpy as np
```

```
print("cos(pi)           = ", np.cos(np.pi))
print('sqaure root(1.21) = ',np.sqrt(1.21))
print(np.log(np.exp(5.2)))
```

```
#List in Arrays
list = [4, 2, 1 ,6, 7]

#Converting to vectors using numpy
vec = np.array(list)
print(vec.ndim,'Dimensionsal:\n')
print(vec)
print('')

#Converting to matrix using numpy
matrix = np.array([[3, 2, 1], [4, 8, 9], [3, 6, 7]])
print(matrix.ndim,'Dimensional:\n')
print(matrix)

#Transpose of mtrix using numpy
print('\nTranspose:\n\n', matrix.T)
```

```
#Printing 1D array
for i in vec:
  print(i, end = " ")
print('')

#Printing 2D Array
for i in matrix:
  print('')
  for j in i:
    print(j, end = " ")   
print('')
print('')

#Printing 2D Array
for x in np.nditer(matrix):
  print(x, end = " ") 
```

```
#Slicing
print(vec[:])
print(vec[0:])
print(vec[2:5])
print(vec[::2])
print(vec[::-1])
```

```
#Views
view1 = vec.copy()
vec[2] = 0
view2 = vec.copy()
view2[2]= 5

print(vec)
print(view1)
print(view2)
```

```
#where
print(np.where(vec == 6))
print(np.where(vec % 2 != 0)) 
print(np.where(vec % 2 == 0)) 
```

```
#sort and binary search
print(np.sort(vec))
print(vec)
print(np.searchsorted(np.sort(vec), 2))
print(np.searchsorted(np.sort(vec), [2, 6, 9]))
vec = np.sort(vec)
print(vec)
print(np.searchsorted(vec, 7))
```
